```python
import pandas as pd

# ==========================================
# STEP 1: LOAD THE DATASETS
# ==========================================
print("Loading datasets...")
df_company = pd.read_csv("ai_company_adoption.csv.gz")
df_industry = pd.read_csv("ai_industry_summary.csv")
df_country = pd.read_csv("country_ai_index.csv")

# ==========================================
# STEP 1A: CHANGE SOME DATA AND KEEP NEW DATA
# ==========================================
print("\nApplying data updates and saving new dataset files...")

df_company = df_company.copy()
df_industry = df_industry.copy()
df_country = df_country.copy()

# Industry-level updates
df_industry.loc[df_industry["industry"] == "Technology", "avg_ai_adoption_rate"] += 1.25
df_industry.loc[df_industry["industry"] == "Finance", "avg_productivity_change_percent"] += 0.35
df_industry.loc[df_industry["industry"] == "Healthcare", "avg_customer_satisfaction"] += 0.10

# Country-level updates
df_country.loc[df_country["country"] == "India", "digital_maturity_index"] += 1.50
df_country.loc[df_country["country"] == "Singapore", "ai_researchers_per_million"] += 12.00
df_country.loc[df_country["country"] == "USA", "ai_patent_filings_2024"] += 150

# Company-level updates
tech_mask = df_company["industry"] == "Technology"
df_company.loc[tech_mask, "revenue_growth_percent"] += 0.25
df_company.loc[tech_mask, "ai_adoption_rate"] += 0.40

# Save updated source datasets so the new data is preserved
df_company.to_csv("ai_company_adoption_updated.csv.gz", index=False, compression="gzip")
df_industry.to_csv("ai_industry_summary_updated.csv", index=False)
df_country.to_csv("country_ai_index_updated.csv", index=False)

# ==========================================
# STEP 2: STANDARD VALIDATION
# ==========================================
print("\nChecking for missing values and duplicates...")
print(f"Missing values in Company Data: {df_company.isnull().sum().sum()}")
print(f"Duplicate rows in Company Data: {df_company.duplicated().sum()}")

# Drop duplicates just in case
df_company = df_company.drop_duplicates()

# ==========================================
# STEP 3: FEATURE ENGINEERING (CUSTOM METRICS)
# ==========================================
print("\nEngineering new features for the dashboard...")

# 1. Net Job Impact: Jobs Created minus Jobs Displaced
df_company['net_job_impact'] = df_company['jobs_created'] - df_company['jobs_displaced']

# 2. Financial ROI in USD (Millions): Calculating the actual dollar amount of growth
# Formula: Annual Revenue * (Revenue Growth Percent / 100)
df_company['roi_usd_millions'] = df_company['annual_revenue_usd_millions'] * (df_company['revenue_growth_percent'] / 100)

# Round the new financial column to 2 decimal places for a cleaner dashboard
df_company['roi_usd_millions'] = df_company['roi_usd_millions'].round(2)

# ==========================================
# STEP 4: MERGING DATA (OPTIONAL BUT RECOMMENDED)
# ==========================================
# While Power BI can link these tables, merging them in Python creates a 
# single, powerful master dataset that is very easy to analyze.

print("\nMerging datasets into a master file...")
# Merge Country Data (Left join on 'country')
df_master = pd.merge(df_company, df_country, on='country', how='left')

# Merge Industry Data (Left join on 'industry')
# Note: Since the industry columns might overlap slightly, we define suffixes if needed.
df_master = pd.merge(df_master, df_industry, on='industry', how='left')

# ==========================================
# STEP 5: EXPORT THE FINAL CLEANED DATASET
# ==========================================
output_filename = "ai_adoption_master_cleaned_updated.csv"
df_master.to_csv(output_filename, index=False)

print("\nUpdated source datasets saved as:")
print("- ai_company_adoption_updated.csv.gz")
print("- ai_industry_summary_updated.csv")
print("- country_ai_index_updated.csv")
print(f"\nPipeline Complete! Data successfully exported as '{output_filename}'")
print(f"Final Dataset Shape: {df_master.shape}")
```
